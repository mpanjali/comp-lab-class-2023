```python
import matplotlib.pyplot as plt
import MDAnalysis as mda
import MDAnalysis.analysis.rdf as rdf
```

```python
Water_energy = "/home/mp6576/comp-class-2023/comp-lab-class-2023/Week12/Runs/Water/water_aimd_1ns_every0.5fs-1.ener"
Water_traj = "/home/mp6576/comp-class-2023/comp-lab-class-2023/Week12/Runs/Water/water_aimd_1ns_every0.5fs-pos-1.xyz"
Water_top = "/home/mp6576/comp-class-2023/comp-lab-class-2023/Week12/Runs/Water/water.xyz"
Proton_energy = "/home/mp6576/comp-class-2023/comp-lab-class-2023/Week12/Runs/Proton/proton_aimd_1ns_every0.5fs-1.ener"
Proton_traj = "/home/mp6576/comp-class-2023/comp-lab-class-2023/Week12/Runs/Proton/proton_aimd_1ns_every0.5fs-pos-1.xyz"
Proton_top = "/home/mp6576/comp-class-2023/comp-lab-class-2023/Week12/Runs/Proton/proton.xyz"

import numpy as np

box = 12.42

Water_energy = np.loadtxt(Water_energy)
Water = mda.Universe(Water_top, Water_traj)
Water.dimensions= np.array([box,box,box,90,90,90])
Proton_energy = np.loadtxt(Proton_energy)
Proton = mda.Universe(Proton_traj, Proton_traj)
Proton.dimensions= np.array([box,box,box,90,90,90])
```

```python
# figure 1 - kinetic energy versus time
fig, ax = plt.subplots(figsize=(5,4))

# plot kinetic energy
ax.plot(Water_energy[:,1]/1000, Water_energy[:,2], color="blue", label='Water')
ax.plot(Proton_energy[:,1]/1000, Proton_energy[:,2], color="red", label='Proton')
# title and labels
ax.set_title('Kinetic Energy vs Time')
ax.set_xlabel('Time (ps)',fontsize=12)
ax.set_ylabel('Kinetic Energy (a.u.)',fontsize=12)
# legend
ax.legend(fontsize=12, frameon=False)

# save figure as pdf in ../Figures
directory = "/home/mp6576/comp-class-2023/comp-lab-class-2023/Week12/Figures/"
fname = "Week12Fig.1-Kinetic Energy.pdf"
full_dir = directory + fname

try:
    plt.savefig(full_dir, bbox_inches='tight')  # Adjust bounding box to fit the plot
    print("Plot saved to PDF successfully.")
except Exception as e:
    print(f"Error saving plot: {e}")
    print("File not saved")
```

```python
# figure 2 - potential energy versus time
fig, ax = plt.subplots(figsize=(5,4))

# plot potential energy
ax.plot(Water_energy[:,1]/1000, Water_energy[:,4], color="blue", label='Water')
ax.plot(Proton_energy[:,1]/1000, Proton_energy[:,4], color="red", label='Proton')
# title and labels
ax.set_title('Potential Energy vs Time')
ax.set_xlabel('Time (ps)',fontsize=12)
ax.set_ylabel('Potential Energy (a.u.)',fontsize=12)
# legend
ax.legend(fontsize=12, frameon=False)

# save figure as pdf in ../Figures
directory = "/home/mp6576/comp-class-2023/comp-lab-class-2023/Week12/Figures/"
fname = "Week12Fig.2-Potential Energy.pdf"
full_dir = directory + fname

try:
    plt.savefig(full_dir, bbox_inches='tight')  # Adjust bounding box to fit the plot
    print("Plot saved to PDF successfully.")
except Exception as e:
    print(f"Error saving plot: {e}")
    print("File not saved")
```

```python
Water_oxygen = Water.select_atoms('name O')
Water_g = rdf.InterRDF(Water_oxygen, Water_oxygen, nbins=100, range=(0.0,box/2), verbose=True)
Water_g.run()
Proton_oxygen = Proton.select_atoms('name O')
Proton_g = rdf.InterRDF(Proton_oxygen,Proton_oxygen,nbins=100, range=(0.0,box/2), verbose=True)
Proton_g.run()
```

```python
# Create figure and axis objects
fig, ax = plt.subplots(figsize=(5, 4))

# Plot g(r)
ax.plot(Water_g.bins[1:], Water_g.rdf[1:], color="blue", label='Water')

# Title and labels
ax.set_title('g(r) vs. r')
ax.set_xlabel('r (Å)', fontsize=12)
ax.set_ylabel('g(r)', fontsize=12)

# Legend
ax.legend(fontsize=12, frameon=False)

# Save figure as PDF
directory = "/home/mp6576/comp-class-2023/comp-lab-class-2023/Week12/Figures/"
fname = "Week12Fig.3-g(r) for Water.pdf"
full_dir = directory + fname

try:
    plt.savefig(full_dir, bbox_inches='tight')  # Adjust bounding box to fit the plot
    print("Plot saved to PDF successfully.")
except Exception as e:
    print(f"Error saving plot: {e}")
    print("File not saved")
```
