```python
import mdtraj as md

#Opening the file
trajectory = md.load('1hz3_T310.stepid25000000.every100ps.nowater.xtc', top='1hz3_T310.start.nowater.gro')

#Calculating the total number of atoms from the opened file
num_atoms = trajectory.n_atoms

#Establishing what the first and last atoms are from the file
first_atom_index = 0
last_atom_index = num_atoms - 1

#Pairing the first and last atoms together
atom_pairs = [[first_atom_index, last_atom_index]]

#Remember that end-to-end distances in a two-dimensional space is computed as the square root of y2-y1/x2-x1. This function essential applies that function with the atom pairs. 
end_to_end_distances = md.compute_distances(trajectory, atom_pairs)

#Print the end-to-end distances for each frame. A frame is a "picture" of a simulation at a particular point in time. 
for frame, distance in enumerate(end_to_end_distances):
    print(f"Frame {frame + 1}: End-to-End Distance = {distance[0]} nm")
```

```python
rg_values = md.compute_rg(trajectory)

# Print the radius of gyration for each frame
for frame, rg in enumerate(rg_values):
    print(f"Frame {frame + 1}: Radius of Gyration = {rg} nm")
```

```python
import matplotlib.pyplot as plt
import numpy as np

rg_interpolated = np.interp(trajectory.time, trajectory.time, rg_values)

#Plot the distribution
fig, ax1 = plt.subplots()

ax1.plot(trajectory.time/1000, end_to_end_distances, color='blue', linestyle='solid', linewidth=2)
ax1.set_xlabel('Time (ns)')
ax1.set_ylabel('End_to_End Distances (nm)', color='blue')

ax2=ax1.twinx()
ax2.plot(trajectory.time/1000,rg_interpolated,color='red', linestyle='dashed', linewidth=2)
ax2.set_ylabel('Radius of Gyration (nm)', color='red')
fig.legend(['End_to_End Distance', 'Radius of Gyration'], loc='upper right', bbox_to_anchor=(1,1),bbox_transform=ax1.transAxes,frameon=False)

plt.show()

# Define the directory and filename
directory = "/home/mp6576/comp-class-2023/comp-lab-class-2023/Week2/Analysis/FinalPlots"
fname = "Week2Part6-Fig.1-End_to_EndDistances(nm)andRadiusofGyration(nm)vsTime(ns).png"

import os
os.chdir("/home/mp6576/comp-class-2023/comp-lab-class-2023/Week2/Data/")

# Check if the directory exists, if not create it
if not os.path.exists(directory):
    try:
        os.makedirs(directory)
    except OSError as e:
        print(f"Error creating directory: {e}")
        print("File not saved")
    else:
        print("Directory created successfully")

# Save the plot to a PNG file
try:
    plt.savefig(os.path.join(directory, fname))
    print("Plot saved to PNG successfully.")
except Exception as e:
    print(f"Error saving plot: {e}")
    print("File not saved")

#Note: plot did not appear in PDF or PNG when saved in "FinalPlots" folder so saved and uploaded separately as PNG
```
