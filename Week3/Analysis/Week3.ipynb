```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

# Load data from file
data = np.loadtxt('/home/mp6576/comp-class-2023/comp-lab-class-2023/Week3/Data/potential.xvg', skiprows=24)

# Clip the data to include values up to 700 ps
data_clipped = data[data[:, 0] <= 720]  # Assuming the first column represents time

# Create a new figure
fig=plt.figure()

# Plot the data
plt.plot(data_clipped[:, 0], data_clipped[:, 1], color='blue', label='Potential', linestyle='-', linewidth=1)

# Add text annotations
plt.text(x=0.5, y=0.88, s="Energy Minimization", fontsize=18, ha="center", transform=fig.transFigure)
plt.text(x=0.5, y=0.82, s="1AKI, Steepest Descent", fontsize=11, ha="center", transform=fig.transFigure)
plt.subplots_adjust(top=0.8, wspace=0.3)

# Set labels and title
plt.xlabel('Energy Minimization Step (ps)', fontsize=14)
plt.ylabel('Potential Energy (kJ/mol)', fontsize=14)

# Set x-axis limits and ticks
plt.xlim(left=0, right=800)
plt.xticks(np.arange(0, 810, 200))
plt.minorticks_on()
minor_locator = MultipleLocator(100)
plt.gca().xaxis.set_minor_locator(minor_locator)

# Format y-axis ticks
plt.ticklabel_format(style='sci', axis='y', scilimits=(0, 0), useMathText=True)

# Add legend
plt.legend(loc='upper right', fontsize=12)

# Define the directory and filename
directory = "/home/mp6576/comp-class-2023/comp-lab-class-2023/Week3/Figures/"
fname = "Week3Tutorial-Fig.1: Energy Minimization.pdf"
full_dir = directory + fname
print(full_dir)
plt.savefig(full_dir)

try:
    plt.savefig(full_dir)
    print("Plot saved to PDF successfully.")
except Exception as e:
    print(f"Error saving plot: {e}")
    print("File not saved")
```

```python
data = '/home/mp6576/comp-class-2023/comp-lab-class-2023/Week3/Data/pressure.xvg'

# Open the file, read lines, and filter out lines starting with '#' and '@'
filtered_lines = []
with open(data, 'r') as file:
    for line in file:
        if not line.startswith(('#', '@')):
            filtered_lines.append(line)

# Now you can process the filtered lines as needed
# For example, if you want to parse the data using NumPy:
import numpy as np
 
# Parse the filtered lines as tab-separated data
data_final = np.loadtxt(filtered_lines)

import matplotlib.pyplot as plt
 
# # Sample data
x = data_final[:, 0]  # Column 1 (assuming zero-based indexing)
y = data_final[:, 1]  # Column 0

# Calculate the running average
window_size = 11  # Adjust the window size as needed
running_average = np.convolve(y, np.ones(window_size)/window_size, mode='valid')
running_average[85:91] = np.nan
 
# Plot the original line plot
plt.plot(x, y)
 
# Plot the running average
plt.plot(x[window_size-1:], running_average, color='red', label='10-ps Running Avg')
 
# Add grid lines
# plt.grid(True)
 
# Set tick interval
# plt.xticks(np.arange(min(x), max(x)+1, 10))
# plt.yticks(np.arange(min(y), max(y)+1, 10))
plt.xlim(left=0, right=100)
 
# Add labels and title
plt.xlabel('Time (ps)')
plt.ylabel('Pressure (bar)')
plt.title('Pressure\n 1AKI, NPT Equilibration')
plt.legend()  # Show legend
 
# Show the plot
# plt.show()

# Define the directory and filename
directory = "/home/mp6576/comp-class-2023/comp-lab-class-2023/Week3/Figures/"
fname = "Week3Tutorial-Fig.3: Pressure.pdf"
full_dir = directory + fname
print(full_dir)
plt.savefig(full_dir)
try:
    plt.savefig(full_dir)
    print("Plot saved to PDF successfully.")
except Exception as e:
    print(f"Error saving plot: {e}")
    print("File not saved")
```
