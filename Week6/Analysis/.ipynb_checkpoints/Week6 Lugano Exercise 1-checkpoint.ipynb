```python
import numpy as np

# data path
colvar_1_path = '/home/mp6576/comp-class-2023/comp-lab-class-2023/Week6/Inputs/A/1.adp_basinA_biasfactor10_pace500_height1.2_sigma0.55_metad_10ns.colvar.txt'
hills_1_path = '/home/mp6576/comp-class-2023/comp-lab-class-2023/Week6/Inputs/A/1.adp_basinA_biasfactor10_pace500_height1.2_sigma0.55_metad_10ns.hills.txt'
grid_1_path = '/home/mp6576/comp-class-2023/comp-lab-class-2023/Week6/Inputs/A/1.adp_basinA_biasfactor10_pace500_height1.2_sigma0.55_metad_10ns.grid.dat'
# load data
colvar_1 = np.loadtxt(colvar_1_path)
hills_1 = np.loadtxt(hills_1_path)
grid_1 = np.loadtxt(grid_1_path)
```

```python
import matplotlib.pyplot as plt

fig=plt.figure(3)

# data processing, load time in ns
time_1 = colvar_1[:,0]/1000
phi_1 = colvar_1[:,1]

# plot phi over time, scatter, first 100000 frame(2ns), colorblind friendly
plt.scatter(time_1[:100000], phi_1[:100000], s=1, c='tab:blue')

# set axis range
plt.xlim(-0.05,2.05)
plt.ylim(-np.pi, np.pi)

# labels, font size 12
plt.xlabel('Time (ns)', fontsize=12)
plt.ylabel(r'$\phi$ (rad)', fontsize=12)

# two lines of title
plt.text(x=0.5, y=0.88, s="Time Evolution of Phi", fontsize=18, ha="center", transform=fig.transFigure)
plt.text(x=0.5, y=0.82, s= "ADP, 10-ns Simulation, Bias Factor 10", fontsize=11, ha="center", transform=fig.transFigure)
plt.subplots_adjust(top=0.8, wspace=0.3)

directory = "/home/mp6576/comp-class-2023/comp-lab-class-2023/Week6/Figures/"
fname = "Week6Fig.3-State A Time Evolution of Phi (2 ns).pdf"
full_dir = directory + fname

try:
    plt.savefig(full_dir, bbox_inches='tight')  # Adjust bounding box to fit the plot
    print("Plot saved to PDF successfully.")
except Exception as e:
    print(f"Error saving plot: {e}")
    print("File not saved")
```

```python
fig=plt.figure(4)

# data processing, load time in ns
time_1 = hills_1[:,0]/1000
phi_1 = hills_1[:,1]
height_1 = hills_1[:,3]

# set colorblind-friendly palatte 
colors = ['blue']

# plot height over time, scatter, colorblind friendly
plt.scatter(time_1, height_1, s=2, c=colors[0])

# set axis range
plt.xlim(-0.08,10.05)
plt.ylim(-0.01, 1.42)

# labels, font size 12
plt.xlabel('Time (ns)', fontsize=12)
plt.ylabel('Gaussian Height (kJ/mol)', fontsize=12)

# two lines of title
plt.text(x=0.5, y=0.88, s="Time Evolution of Gaussian Height", fontsize=18, ha="center", transform=fig.transFigure)
plt.text(x=0.5, y=0.82, s= "ADP, 10-ns Simulation, Bias Factor 10", fontsize=11, ha="center", transform=fig.transFigure)
plt.subplots_adjust(top=0.8, wspace=0.3)

directory = "/home/mp6576/comp-class-2023/comp-lab-class-2023/Week6/Figures/"
fname = "Week6Fig.4-State A Time Evolution of Gaussian Height.pdf"
full_dir = directory + fname

try:
    plt.savefig(full_dir, bbox_inches='tight')  # Adjust bounding box to fit the plot
    print("Plot saved to PDF successfully.")
except Exception as e:
    print(f"Error saving plot: {e}")
    print("File not saved")
```
