```python
crystal_structure = '/home/mp6576/comp-class-2023/comp-lab-class-2023/Week7/Inputs/1ubq.pdb'

colab_structure_0 = '/home/mp6576/comp-class-2023/comp-lab-class-2023/Week7/Inputs/AlphaFold Colab/test_7686d_0_unrelaxed_rank_001_alphafold2_ptm_model_4_seed_000.pdb'
colab_structure_1 = '/home/mp6576/comp-class-2023/comp-lab-class-2023/Week7/Inputs/AlphaFold Colab/test_7686d_0_unrelaxed_rank_002_alphafold2_ptm_model_3_seed_000.pdb'
colab_structure_2 = '/home/mp6576/comp-class-2023/comp-lab-class-2023/Week7/Inputs/AlphaFold Colab/test_7686d_0_unrelaxed_rank_003_alphafold2_ptm_model_2_seed_000.pdb'
colab_structure_3 = '/home/mp6576/comp-class-2023/comp-lab-class-2023/Week7/Inputs/AlphaFold Colab/test_7686d_0_unrelaxed_rank_004_alphafold2_ptm_model_1_seed_000.pdb'
colab_structure_4 = '/home/mp6576/comp-class-2023/comp-lab-class-2023/Week7/Inputs/AlphaFold Colab/test_7686d_0_unrelaxed_rank_005_alphafold2_ptm_model_5_seed_000.pdb'

greene_structure_0 = '/home/mp6576/comp-class-2023/comp-lab-class-2023/Week7/Inputs/rcsb_pdb_1UBQ_AlphaFold Greene/unrelaxed_model_1_pred_0.pdb'
greene_structure_1 = '/home/mp6576/comp-class-2023/comp-lab-class-2023/Week7/Inputs/rcsb_pdb_1UBQ_AlphaFold Greene/unrelaxed_model_2_pred_0.pdb'
greene_structure_2 = '/home/mp6576/comp-class-2023/comp-lab-class-2023/Week7/Inputs/rcsb_pdb_1UBQ_AlphaFold Greene/unrelaxed_model_3_pred_0.pdb'
greene_structure_3 = '/home/mp6576/comp-class-2023/comp-lab-class-2023/Week7/Inputs/rcsb_pdb_1UBQ_AlphaFold Greene/unrelaxed_model_4_pred_0.pdb'
greene_structure_4 = '/home/mp6576/comp-class-2023/comp-lab-class-2023/Week7/Inputs/rcsb_pdb_1UBQ_AlphaFold Greene/unrelaxed_model_5_pred_0.pdb'
```

```python
import mdtraj as md
import numpy as np

crystal_pdb = md.load_pdb(crystal_structure)
crystal_pdb = crystal_pdb.atom_slice(crystal_pdb.top.select('chainid 0'))

greene_0_pdb = md.load_pdb(greene_structure_0)
greene_1_pdb = md.load_pdb(greene_structure_1)
greene_2_pdb = md.load_pdb(greene_structure_2)
greene_3_pdb = md.load_pdb(greene_structure_3)
greene_4_pdb = md.load_pdb(greene_structure_4)

colab_0_pdb = md.load_pdb(colab_structure_0)
colab_1_pdb = md.load_pdb(colab_structure_1)
colab_2_pdb = md.load_pdb(colab_structure_2)
colab_3_pdb = md.load_pdb(colab_structure_3)
colab_4_pdb = md.load_pdb(colab_structure_4)
```

```python
selection_1 = 'name CA'
# align the structure
crystal_pdb = crystal_pdb.superpose(crystal_pdb, 0, atom_indices=crystal_pdb.top.select(selection_1))
greene_0_pdb = greene_0_pdb.superpose(crystal_pdb, 0, atom_indices=crystal_pdb.top.select(selection_1))
greene_1_pdb = greene_1_pdb.superpose(crystal_pdb, 0, atom_indices=crystal_pdb.top.select(selection_1))
greene_2_pdb = greene_2_pdb.superpose(crystal_pdb, 0, atom_indices=crystal_pdb.top.select(selection_1))
greene_3_pdb = greene_3_pdb.superpose(crystal_pdb, 0, atom_indices=crystal_pdb.top.select(selection_1))
greene_4_pdb = greene_4_pdb.superpose(crystal_pdb, 0, atom_indices=crystal_pdb.top.select(selection_1))
colab_0_pdb = colab_0_pdb.superpose(crystal_pdb, 0, atom_indices=crystal_pdb.top.select(selection_1))
colab_1_pdb = colab_1_pdb.superpose(crystal_pdb, 0, atom_indices=crystal_pdb.top.select(selection_1))
colab_2_pdb = colab_2_pdb.superpose(crystal_pdb, 0, atom_indices=crystal_pdb.top.select(selection_1))
colab_3_pdb = colab_3_pdb.superpose(crystal_pdb, 0, atom_indices=crystal_pdb.top.select(selection_1))
colab_4_pdb = colab_4_pdb.superpose(crystal_pdb, 0, atom_indices=crystal_pdb.top.select(selection_1))

# calculate RMSD on alpha carbon
greene_0_rmsd_carbon = md.rmsd(crystal_pdb, greene_0_pdb, atom_indices=crystal_pdb.top.select(selection_1))
greene_1_rmsd_carbon = md.rmsd(crystal_pdb, greene_1_pdb, atom_indices=crystal_pdb.top.select(selection_1))
greene_2_rmsd_carbon = md.rmsd(crystal_pdb, greene_2_pdb, atom_indices=crystal_pdb.top.select(selection_1))
greene_3_rmsd_carbon = md.rmsd(crystal_pdb, greene_3_pdb, atom_indices=crystal_pdb.top.select(selection_1))
greene_4_rmsd_carbon = md.rmsd(crystal_pdb, greene_4_pdb, atom_indices=crystal_pdb.top.select(selection_1))
colab_0_rmsd_carbon = md.rmsd(crystal_pdb, colab_0_pdb, atom_indices=crystal_pdb.top.select(selection_1))
colab_1_rmsd_carbon = md.rmsd(crystal_pdb, colab_1_pdb, atom_indices=crystal_pdb.top.select(selection_1))
colab_2_rmsd_carbon = md.rmsd(crystal_pdb, colab_2_pdb, atom_indices=crystal_pdb.top.select(selection_1))
colab_3_rmsd_carbon = md.rmsd(crystal_pdb, colab_3_pdb, atom_indices=crystal_pdb.top.select(selection_1))
colab_4_rmsd_carbon = md.rmsd(crystal_pdb, colab_4_pdb, atom_indices=crystal_pdb.top.select(selection_1))

selection_2 = 'backbone'
# calculate RMSD on backbone
greene_0_rmsd_backbone = md.rmsd(crystal_pdb, greene_0_pdb, atom_indices=crystal_pdb.top.select(selection_2))
greene_1_rmsd_backbone = md.rmsd(crystal_pdb, greene_1_pdb, atom_indices=crystal_pdb.top.select(selection_2))
greene_2_rmsd_backbone = md.rmsd(crystal_pdb, greene_2_pdb, atom_indices=crystal_pdb.top.select(selection_2))
greene_3_rmsd_backbone = md.rmsd(crystal_pdb, greene_3_pdb, atom_indices=crystal_pdb.top.select(selection_2))
greene_4_rmsd_backbone = md.rmsd(crystal_pdb, greene_4_pdb, atom_indices=crystal_pdb.top.select(selection_2))
colab_0_rmsd_backbone = md.rmsd(crystal_pdb, colab_0_pdb, atom_indices=crystal_pdb.top.select(selection_2))
colab_1_rmsd_backbone = md.rmsd(crystal_pdb, colab_1_pdb, atom_indices=crystal_pdb.top.select(selection_2))
colab_2_rmsd_backbone = md.rmsd(crystal_pdb, colab_2_pdb, atom_indices=crystal_pdb.top.select(selection_2))
colab_3_rmsd_backbone = md.rmsd(crystal_pdb, colab_3_pdb, atom_indices=crystal_pdb.top.select(selection_2))
colab_4_rmsd_backbone = md.rmsd(crystal_pdb, colab_4_pdb, atom_indices=crystal_pdb.top.select(selection_2))

selection_3 = 'sidechain and not (index == 601)'
greene_0_rmsd_sidechain = md.rmsd(crystal_pdb, greene_0_pdb, atom_indices=crystal_pdb.top.select(selection_3))
greene_1_rmsd_sidechain = md.rmsd(crystal_pdb, greene_1_pdb, atom_indices=crystal_pdb.top.select(selection_3))
greene_2_rmsd_sidechain = md.rmsd(crystal_pdb, greene_2_pdb, atom_indices=crystal_pdb.top.select(selection_3))
greene_3_rmsd_sidechain = md.rmsd(crystal_pdb, greene_3_pdb, atom_indices=crystal_pdb.top.select(selection_3))
greene_4_rmsd_sidechain = md.rmsd(crystal_pdb, greene_4_pdb, atom_indices=crystal_pdb.top.select(selection_3))
colab_0_rmsd_sidechain = md.rmsd(crystal_pdb, colab_0_pdb, atom_indices=crystal_pdb.top.select(selection_3))
colab_1_rmsd_sidechain = md.rmsd(crystal_pdb, colab_1_pdb, atom_indices=crystal_pdb.top.select(selection_3))
colab_2_rmsd_sidechain = md.rmsd(crystal_pdb, colab_2_pdb, atom_indices=crystal_pdb.top.select(selection_3))
colab_3_rmsd_sidechain = md.rmsd(crystal_pdb, colab_3_pdb, atom_indices=crystal_pdb.top.select(selection_3))
colab_4_rmsd_sidechain = md.rmsd(crystal_pdb, colab_4_pdb, atom_indices=crystal_pdb.top.select(selection_3))
```

```python
fig = plt.figure(figsize=(10, 5))

# data set
data = [greene_0_rmsd_carbon, greene_1_rmsd_carbon, greene_2_rmsd_carbon, greene_3_rmsd_carbon, greene_4_rmsd_carbon, colab_0_rmsd_carbon, colab_1_rmsd_carbon, colab_2_rmsd_carbon, colab_3_rmsd_carbon, colab_4_rmsd_carbon]
# data set labels
labels = ['Greene_0', 'Greene_1', 'Greene_2', 'Greene_3', 'Greene_4', 'Colab_0', 'Colab_1', 'Colab_2', 'Colab_3', 'Colab_4']

# plot bars
bar_width = 0.8
for i in range(len(data)):
    plt.bar(labels[i], data[i], width=bar_width, color='blue', edgecolor='black')
    i += 1

# labels, font size 12
plt.xlabel('Methods', fontsize=12)
plt.ylabel('RMSD (nm)', fontsize=12)

# x ticks, font size 8
plt.xticks(fontsize=8)

# two lines of title
plt.text(x=0.5, y=0.85, s="RMSD: Alpha Carbons", fontsize=18, ha="center", transform=fig.transFigure)
plt.text(x=0.5, y=0.82, s= "Ubiquitin, Unrelaxed Structures vs Crystal Structure", fontsize=11, ha="center", transform=fig.transFigure)
plt.subplots_adjust(top=0.8, wspace=0.3)

# save figure as pdf in ../Figures
directory = "/home/mp6576/comp-class-2023/comp-lab-class-2023/Week7/Figures/"
fname = "Week7Fig.2-RMSD Alpha Carbons (Ubiquitin).pdf"
full_dir = directory + fname

try:
    plt.savefig(full_dir, bbox_inches='tight')  # Adjust bounding box to fit the plot
    print("Plot saved to PDF successfully.")
except Exception as e:
    print(f"Error saving plot: {e}")
    print("File not saved")
```
