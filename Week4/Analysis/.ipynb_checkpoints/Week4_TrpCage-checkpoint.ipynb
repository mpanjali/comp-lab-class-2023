```python
import mdtraj as md

xtc_file='Trp-Cage_center.xtc'
gro_file='Trp-Cage_center.gro'
traj=md.load(xtc_file,top=gro_file)
```

```python
import numpy as np
import matplotlib.pyplot as plt

end_to_end_distances=md.compute_distances(traj,[[0,1]])

fig = plt.figure(figsize=(8, 6))  # Adjust the figure size as needed
plt.plot(traj.time / 1000, end_to_end_distances, color='#0072B2')
plt.xlabel('Time (ns)', fontsize=12)
plt.ylabel('End_to_End Distances (nm)', fontsize=12)
plt.text(x=0.5, y=0.88, s="End_to_End Distances vs. Time", fontsize=18, ha="center", transform=fig.transFigure)
plt.text(x=0.5, y=0.82, s="TrpCage_AMBER", fontsize=11, ha="center", transform=fig.transFigure)
plt.subplots_adjust(top=0.8, wspace=0.3)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlim(0, 100)
plt.yticks(np.arange(0.1, 0.103, 0.0005))

directory = "/home/mp6576/comp-class-2023/comp-lab-class-2023/Week4/Figures/TrpCage_AMBER/"
fname = "Week4Fig.1-End-to-End_Distances.pdf"
full_dir = directory + fname
print(full_dir)
plt.savefig(full_dir, bbox_inches='tight')

try:
    plt.savefig(full_dir)
    print("Plot saved to PDF successfully.")
except Exception as e:
    print(f"Error saving plot: {e}")
    print("File not saved")
```

```python
rg_values=md.compute_rg(traj)
rg_interpolated = np.interp(traj.time, traj.time, rg_values)

fig=plt.figure(2)
plt.plot(traj.time/1000,rg_interpolated,color='#0072B2')
# set the axis labels, fontsize = 12
plt.xlabel('Time (ns)',fontsize=12)
plt.ylabel('Radius of Gyration (nm)',fontsize=12)
# two lines of title
plt.text(x=0.5, y=0.88, s="Radius of Gyration vs. Time", fontsize=18, ha="center", transform=fig.transFigure)
plt.text(x=0.5, y=0.82, s= "TrpCage_AMBER", fontsize=11, ha="center", transform=fig.transFigure)
plt.subplots_adjust(top=0.8, wspace=0.3)
# modify the tick labels to be fontsize = 12
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
# set limits for the axes
plt.xlim(0,100)

directory = "/home/mp6576/comp-class-2023/comp-lab-class-2023/Week4/Figures/TrpCage_AMBER/"
fname = "Week4Fig.2-Radius of Gyration.pdf"
full_dir = directory + fname
print(full_dir)
plt.savefig(full_dir, bbox_inches='tight')

try:
    plt.savefig(full_dir)
    print("Plot saved to PDF successfully.")
except Exception as e:
    print(f"Error saving plot: {e}")
    print("File not saved")
```

```python
fig = plt.figure(figsize=(8, 6))  # Adjust the figure size as needed
ax1 = fig.add_subplot(111)
ax1.plot(traj.time / 1000, end_to_end_distances, color='#0072B2')
ax1.set_xlabel('Time (ns)', fontsize=12)
ax1.set_ylabel('End_to_End Distances (nm)', fontsize=12)
plt.text(x=0.5, y=0.88, s="End_to_End Distances and Radius of Gyration", fontsize=18, ha="center", transform=fig.transFigure)
plt.text(x=0.5, y=0.82, s="TrpCage_AMBER", fontsize=11, ha="center", transform=fig.transFigure)
plt.subplots_adjust(top=0.8, wspace=0.3)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlim(0, 100)
plt.ylim(0.1, 0.103)
plt.yticks(np.arange(0.1, 0.1031, 0.0005))

ax2 = ax1.twinx()
ax2.plot(traj.time / 1000, rg_interpolated, color='#D55E00')
ax2.set_ylabel('Radius of Gyration (nm)', fontsize=12)

# Adjust the legend location to fit both plots
fig.legend(['End_to_End Distances (nm)', 'Radius of Gyration (nm)'], loc='upper right', bbox_to_anchor=(1, 1), bbox_transform=ax1.transAxes, frameon=False)

# Save figure as PDF
directory = "/home/mp6576/comp-class-2023/comp-lab-class-2023/Week4/Figures/TrpCage_AMBER/"
fname = "Week4Fig.3-End_to_End Distances and Radius of Gyration.pdf"
full_dir = directory + fname

try:
    plt.savefig(full_dir, bbox_inches='tight')  # Adjust bounding box to fit the plot
    print("Plot saved to PDF successfully.")
except Exception as e:
    print(f"Error saving plot: {e}")
    print("File not saved")
```
