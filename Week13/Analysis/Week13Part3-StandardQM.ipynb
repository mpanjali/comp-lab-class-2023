```python
import numpy as np

energymonitor_file = "/home/mp6576/comp-class-2023/comp-lab-class-2023/Week13/Outputs/monitor/MONITOR-COLVAR.metadynLog"
energymonitor = np.loadtxt(energymonitor_file,skiprows=1)
energymd_file = "/home/mp6576/comp-class-2023/comp-lab-class-2023/Week13/Outputs/metad1_initmonitor/METADYN-COLVAR.metadynLog"
energymd = np.loadtxt(energymd_file,skiprows=1)
energy_mdfh_file = "/home/mp6576/comp-class-2023/comp-lab-class-2023/Week13/Outputs/metad1_initmonitor_freqhills/METADYN-COLVAR.metadynLog"
energy_mdfh = np.loadtxt(energy_mdfh_file, skiprows=1)
```

```python
import matplotlib.pyplot as plt

def plot_data(x, y, label, color, linewidth):
    plt.plot(x / 1000, y, color=color, label=label, linewidth=linewidth)

def set_plot_properties():
    plt.xlabel('Time (ps)', fontsize=12)
    plt.ylabel('Reaction coordinate $d_1 - d_2$ (nm)', fontsize=12)
    plt.xlim(0, 30)
    plt.ylim(-20, 11)
    plt.axhline(y=0, color='#BBBBBB', linestyle='--', linewidth=1)
    plt.legend(loc='lower right', fontsize=12, frameon=False)
    plt.title('Reaction Coordinate vs. Time', fontsize=14) 
    plt.figure(6)

# set colorblind-friendly palette
colors = ['blue', 'red', 'green']
line_width = [2, 1.5, 1]

# define data
data = [
    (energymonitor[190:, 0], energymonitor[190:, 1], 'Standard QM/MM', colors[0], line_width[0]),
    (energymd[:, 0], energymd[:, 1], 'Metadynamics - Freqhills ', colors[1], line_width[1]),
    (energy_mdfh[:, 0], energy_mdfh[:, 1], 'Metadynamics - Freqhills', colors[2], line_width[2])
]

# plot each data set
for x, y, label, color, linewidth in data:
    plot_data(x, y, label, color, linewidth)

# set plot properties
set_plot_properties()

# save figure as pdf in ../Figures
directory = "/home/mp6576/comp-class-2023/comp-lab-class-2023/Week13/Figures/"
fname = "Week13Fig.6-Reaction Coordinate Evolution.pdf"
full_dir = directory + fname

try:
    plt.savefig(full_dir, bbox_inches='tight')  # Adjust bounding box to fit the plot
    print("Plot saved to PDF successfully.")
except Exception as e:
    print(f"Error saving plot: {e}")
    print("File not saved")
```
