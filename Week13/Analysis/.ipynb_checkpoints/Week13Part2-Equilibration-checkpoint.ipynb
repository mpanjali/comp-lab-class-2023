```python
import numpy as np

energyNPT_file = "/home/mp6576/comp-class-2023/comp-lab-class-2023/Week13/Outputs/NPT-1.ener"
energyNPT = np.loadtxt(energyNPT_file,skiprows=1)
energyNVT_file = "/home/mp6576/comp-class-2023/comp-lab-class-2023/Week13/Outputs/NVT-1.ener"
energyNVT = np.loadtxt(energyNVT_file,skiprows=1)
energymonitor_file = "/home/mp6576/comp-class-2023/comp-lab-class-2023/Week13/Outputs/monitor/MONITOR-1.ener"
energymonitor = np.loadtxt(energymonitor_file,skiprows=1)
energymd_file = "/home/mp6576/comp-class-2023/comp-lab-class-2023/Week13/Outputs/metad1_initmonitor/METADYN-1.ener"
energymd = np.loadtxt(energymd_file,skiprows=1)
energy_mdfh_file = "/home/mp6576/comp-class-2023/comp-lab-class-2023/Week13/Outputs/metad1_initmonitor_freqhills/METADYN-1.ener"
energy_mdfh = np.loadtxt(energy_mdfh_file, skiprows=1)
```

```python
import matplotlib.pyplot as plt

# figure 1 - potential energy versus time
fig, ax = plt.subplots(figsize=(5,4))

# plot potential energy
ax.plot(energyNPT[:,1]/1000, energyNPT[:,4], color="blue", label='NPT')

# title and labels
ax.set_title('Potential Energy vs Time')
ax.set_xlabel('Time (ps)',fontsize=12)
ax.set_ylabel('Potential Energy (a.u.)',fontsize=12)
# legend
ax.legend(fontsize=12, frameon=False)

# save figure as pdf in ../Figures
directory = "/home/mp6576/comp-class-2023/comp-lab-class-2023/Week13/Figures/"
fname = "Week13Fig.1-Potential Energy for NPT.pdf"
full_dir = directory + fname

try:
    plt.savefig(full_dir, bbox_inches='tight')  # Adjust bounding box to fit the plot
    print("Plot saved to PDF successfully.")
except Exception as e:
    print(f"Error saving plot: {e}")
    print("File not saved")
```
